```python
import pandas as pd
# Leer el primer CSV y obtener los códigos de la segunda columna
df_codigos = pd.read_csv('C:\\Users\\Usuario\\Documents\\Workspace\\Estudio-Psiquiatricos\\datasets\\neo4j\\diagnosticos_psiquiatricos_por_paciente.csv')
codigos = df_codigos.iloc[:, 1]

# Leer el segundo CSV que contiene códigos y descripciones
df_descripciones = pd.read_csv('C:\\Users\\Usuario\\Documents\\Workspace\\Estudio-Psiquiatricos\\datasets\\Conversor_Definitivo.csv', sep='|')
# Crear un diccionario de código a descripción
codigo_a_desc = dict(zip(df_descripciones.iloc[:, 1], df_descripciones.iloc[:, 2]))

# Reemplazar los códigos por las descripciones en el primer DataFrame
df_codigos.iloc[:, 1] = codigos.map(codigo_a_desc)

# Guardar el resultado en un nuevo CSV
df_codigos.to_csv('primer_archivo_con_descripciones.csv', sep='|', index=False)
```
