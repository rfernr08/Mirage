```python
import pandas as pd

# Cargar el CSV original
df = pd.read_csv('C:\\Users\\Usuario\\Documents\\Workspace\\Estudio-Psiquiatricos\\datasets\\PSQ_F20_F29_Ext.csv', sep='|')

# Encontrar el índice de la primera columna que empieza por 'Diag' (ignorando mayúsculas/minúsculas)
first_diag_idx = next(i for i, col in enumerate(df.columns) if col.lower().startswith('diag'))

# Seleccionar solo las columnas antes de la primera columna 'Diag'
cols_no_diag = df.columns[:first_diag_idx]
df_no_diag = df[cols_no_diag]

# Guardar el nuevo CSV sin columnas de diagnóstico ni conjuntos
df_no_diag.to_csv('pacientes_sin_diagnosticos.csv', sep='|', index=False)

# Extraer columnas de diagnóstico
diag_cols = [col for col in df.columns if col.lower().startswith('diag') and not col.startswith('DIAG PSQ')]

# Obtener la lista de diagnósticos para cada paciente (por Nº Historia)
diagnosticos_por_paciente = df.groupby('Nº Historia')[diag_cols].agg(lambda x: x.dropna().tolist()).reset_index()

# Obtener el diagnóstico psiquiátrico principal para cada paciente
diag_psq_por_paciente = df.groupby('Nº Historia')['DIAG PSQ'].agg(lambda x: x.dropna().unique().tolist()).reset_index()

# Mostrar ejemplos
print(diagnosticos_por_paciente.head())
print(diag_psq_por_paciente.head())

diagnosticos_por_paciente.to_csv('diagnosticos_por_paciente.csv', index=False)
diag_psq_por_paciente.to_csv('diagnosticos_psiquiatricos_por_paciente.csv', index=False)
```
