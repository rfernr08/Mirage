```python
import polars as pl
import re

# Leer el archivo Excel
df = pl.read_excel("C:\\Users\\Usuario\\Documents\\Workspace\\Estudio-Psiquiatricos\\datasets\\PSQ_F20_F29_Ext.xlsx")

# Obtener todos los valores Ãºnicos de la columna DIAG PSQ
diagnosticos_combinados = df['DIAG PSQ'].unique().to_list()

# Conjunto para almacenar todos los diagnÃ³sticos Ãºnicos
diagnosticos_unicos = set()

# Procesar cada diagnÃ³stico
for diag in diagnosticos_combinados:
    if diag is not None:  # Verificar que no sea None/null
        diag_str = str(diag)
        # Separar por comas y limpiar espacios
        codigos = [codigo.strip() for codigo in diag_str.split(',')]
        # AÃ±adir cada cÃ³digo al conjunto
        diagnosticos_unicos.update(codigos)

# Convertir a lista ordenada
diagnosticos_lista = sorted(list(diagnosticos_unicos))

# Mostrar resultados
print("ğŸ“‹ DIAGNÃ“STICOS ÃšNICOS ENCONTRADOS:")
print("=" * 50)
for i, diagnostico in enumerate(diagnosticos_lista, 1):
    print(f"{i:2d}. {diagnostico}")

print(f"\nğŸ“Š Total de diagnÃ³sticos Ãºnicos: {len(diagnosticos_lista)}")

# Crear DataFrame con los diagnÃ³sticos Ãºnicos
diagnosticos_df = pl.DataFrame({
    "DIAG_PSQ": diagnosticos_lista
})

# Guardar en CSV
diagnosticos_df.write_csv('diagnosticos_unicos.csv')
print("ğŸ’¾ Archivo guardado: diagnosticos_unicos.csv")

# Mostrar el DataFrame creado
print("\nğŸ“„ DataFrame creado:")
print(diagnosticos_df)
```
